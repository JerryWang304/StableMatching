```python
# 实现stable matching: one-to-one
import numpy
import random
# 男人或女人的个数
num = 5

woman = [i for i in range(5)]
man = [i for i in range(5)]
perference_list_of_man = {}
perference_list_of_woman = {}
# 对每个男人，随机生成一个序列
# random.sample([i for i in range(num)],num)
for m in man:
    pre = numpy.random.permutation(num).tolist()
    perference_list_of_man[m] = pre
    
#对每个女人，也随机生成一个序列
for w in woman:
    pre = numpy.random.permutation(num).tolist()
    perference_list_of_woman[w] = pre
print 'Man: ',   
print perference_list_of_man
print 'Woman: ',
print perference_list_of_woman
freeman = man[:]
engaged = [False for i in range(num)] # 女人是否订婚
married = [0 for i in range(num)]
proposed = numpy.zeros((num,num)) # 求婚矩阵
if __name__ == '__main__':
    while len(freeman) > 0:
        man = freeman[0]
        # 开始求婚: 找到没求过婚的（没有被拒绝过的）
        for w in perference_list_of_man[man]:
            if proposed[man][w] == 0:
                woman = w
                break
        proposed[man][woman] = 1 # 求婚
        if not engaged[woman]:
            engaged[woman] = True
            freeman.remove(man)
            married[woman] = man
        else:
            old = perference_list_of_woman[woman].index(married[woman])
            new = perference_list_of_woman[woman].index(man)
            # 喜新厌旧，水性杨花
            if new < old:
                freeman.remove(man)
                freeman.append(married[woman])
                married[woman] = man
            # 海誓山盟，至死不渝
            else:
                continue
    print "\nwoman  man"
    for woman in range(num):
        print woman,'\t',married[woman]
                
            
```
